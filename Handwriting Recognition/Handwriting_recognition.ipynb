##Load MNIST dataset from Sklearn with fetch_openml
here we use mnist_784. 784 are features(28x28 pixels)

```
from sklearn.datasets import fetch_openml
```

```
mnist=fetch_openml('mnist_784')
```

```
x,y=mnist["data"],mnist["target"]
```

##Check the shape of X and Y(Target)


```
print(x.shape)
print("y shape",y.shape)
```

```
y_int=y.astype(np.int8)
x_int=x.astype(np.int8)  # this is not required for x because it is already in int.
```

#Selecting Models

```
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

# We can use any model 

```
# model=LogisticRegression()
model=DecisionTreeClassifier()
```

```
model.fit(x,y)
```

#test the accuracy 

```
predicted_value=model.predict(x)
```

```
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(predicted_value,y)
print(mse) 
```

#check the accuracy using cross validation 
whether my model is overfitted or not

```
from sklearn.model_selection import cross_val_score
```

```
crossValidation=cross_val_score(model,x,y,cv=10)
print(crossValidation.mean()) 
```
