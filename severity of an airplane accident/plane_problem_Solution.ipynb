```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```
data_plane=pd.read_csv("train.csv")
data_plane
#data_plane.describe()
```

## Lable Encoder for convert String to Integer values


```
from sklearn.preprocessing import LabelEncoder
coded_lable=LabelEncoder()
x=coded_lable.fit_transform(data_plane.iloc[:,0])
data_plane["Severity"]=x
```

```
data_plane.info()
```

```
data_plane.head()
```

```
#data_plane.corr()
data_plane['Accident_ID'].value_counts()         # drop id columns
data_plane.drop(columns=["Accident_ID"],axis=1,inplace=True)
```

```
data_plane.describe()
```

## Model Selection and prediction


```
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
#model=SVC(kernel='linear')
model.fit(data_plane.iloc[:,1:],data_plane.iloc[:,0])
predicted_taregt=model.predict(data_plane.iloc[:100,1:])

```

```
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(predicted_taregt,data_plane.iloc[:100,0])
print(mse)
```

```
from sklearn.model_selection import cross_val_score
score=cross_val_score(model,data_plane.iloc[:,1:],data_plane.iloc[:,0],cv=10)
print("Accuracy=",score.mean())
```

```
test_plane=pd.read_csv("test.csv")
test_plane.describe()
test_plane.drop(columns=['Accident_ID'],axis=1,inplace=True)
model.predict(test_plane)
```
